### Bytes vs String

```python
mystr = "Hello world"
print(type(mystr))
```

```python
mystr = b"Hello world"
print(type(mystr))
print(mystr)
```

### byte values

```python
data = b"ABC"
print(data[0]) #AsCII(A) = 65
```

### Iterating through bytes

```python
data = b"ABC"
for i in data:
    print(i)
```

### Convert string → bytes

```python
#Method 1
text = "Hello"
byte = text.encode()
print(byte)
```

```python
#Method 2
text = "Hello"
print(bytes(text, "utf-8"))
```

### convert bytes -> string

```python
text = byte.decode()
print(text)
```

```python
#Method 1
byte = b"Hello"
text = byte.decode()
print(text)
```

```python
#We cannot simply use str() it wont work
byte = b"Hello"
print(str(byte))
```

```python
#we can do it as -> Method 2
byte = b"Hello"
print(str(byte, "utf-8"))
```

```python
data = b'Python'
print(len(data))
```

```python
#string -> bytes can be done using 'ascii' too
text = "Hello"
print(bytes(text, 'utf-8'))
print(text)
print(bytes(text, "ascii"))
```

### Zero-filled bytes object of a specified length:

```python
print("bytes(10): ", bytes(10))
```

### Immutability

```python
text = "Hello World"
print(text[6])
```

```python
text[6] = 'p'
print(text[6])
```

```python
print(text.replace('W', 'p'))
print(text)
```
