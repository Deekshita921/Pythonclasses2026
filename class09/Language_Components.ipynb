## Relational Operations
    These operations will result in boolean value
    (>, <, ==, >=, <=, !=)

```python
a = 20
b = 10
print(f"{a > b = }")
print(f"{a < b =}")
print(f"{a == b =}")
print(f"{a >= b =}")
print(f"{a <= b =}")
print(f"{a != b =}")
```

```python
print(f"a != b = {a!=b}")
```

```python
str1 = "abc"
str2 = "abc"
print(f"{str1 == str2 =}")
print(f"{str1 > str2 =}")
print(f"{str1 >= str2 =}")
print(f"{str1 < str2 =}")
print(f"{str1 <= str2 =}")
print(f"{str1 != str2 =}")
print(f"str1 != str2 = {str1!=str2}")
```

```python
str1 = "abc"
str2 = "bc"
print(f"{str1 == str2 =}")
print(f"{str1 > str2 =}")
print(f"{str1 >= str2 =}")
print(f"{str1 < str2 =}")
print(f"{str1 <= str2 =}")
print(f"{str1 != str2 =}")
print(f"str1 != str2 = {str1!=str2}")
```

## Logical Operations
    will result in boolean values
    (and, or, not)

```python
print(True and True) #True
print(True and False) #false
print(False and True) #False
print(False and False) #False
print(True and False and False)
```

```python
print(True or True) 
print(True or False) 
print(False or True) 
print(False or False) 
print(True or False or False)
```

```python
print(not True)
print(not False)
print(not not True) #not not True->not Flase
print(not not False)
```
