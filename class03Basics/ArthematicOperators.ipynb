##Arithmetic Operations

```python
num1 = 10
num2 = 20
```

```python
print(num1 + num2)
```

```python
print(num1 - num2)
```

```python
print(num1 * num2)
```

```python
print(num1 / num2)
print(10/3)
```

```python
print(num1 // num2) #This is called floor divison which round ups the result value
print(10//3)
```

```python
print(num1 % num2) #num1 = 10, num2 = 20
print(4%2) #answer=0. Because 4 is divided by 2. 4/2 remainder=0
```

```python
print(2**2) #which means 2 square = 4
print(2**3) #which means 2 power 3 = 8
```

#Power Operator (**)

```python
4 ** 2
```

```python
pow(4,2)
```

```python
64 ** (1/2) #square root function
```

```python
pow(4,2) == 4 ** 2
```

```python
pow(64, 1/2) == 64 ** (1/2) #square root 
```

```python
pow(64, 0.5) == 64 ** 0.5
```

To do both div and mod using pow() operator

```python
pow(4,2,9) #means for 1st 2 it do pow operation and for next num it do module for the answer of pow
#pow(4,2) = 16 and 16%9=7
```

```python
pow(4,2,9) == 4 ** 2 % 9
```

#Compound Operators

```python
temp = 0
temp = temp + 1
print(temp)
```

```python
temp = 0
temp += 1
print(temp)
```

```python
a = 1; temp = 1
temp = temp + a + 1 
print(temp)
```

```python
a = 1; temp = 1
temp += a + 1
print(temp)
```
