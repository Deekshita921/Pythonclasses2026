```python
print('Today is an awesome day \
to learn python')
```

```python
print("Today is an awesome day \
to learn python")
```

```python
print('''Today is an awesome day \
to learn python''')
```

```python
print('''Hello,
      Welcome to python
      Learn Strings''') #If we give space then it will also will be there in output
```

```python
print('''Hello,
Welcome to python
Learn Strings''')
```

```python
print("""Hello,
Welcome to python
Learn Strings""")
```

## String Slicing Operations

```python
#indexing
print(word[0])
```

```python
#Slicing
print(word[0:11])
```

```python
print(word[5:17]) #python doesn't include the last value, in a boundary condition.
print(word[7:10])
```

```python
print(word[0:5])
print(word[0:6])
```

```python
print(word[18])
```

```python
print("word[0:18]:",word[0:18]) #programming
print("word[7:999]:",word[7:999]) #999 index isn't present
print("word[45:87]:",word[45:87]) #indexes are not present
```

```python
word = "python"
print(word[0:3]) #pyt
print(word[2:]) #thon. It goes 2 till end
print(word[:4]) #pyth. It assumes start=0
print(word[:]) #python. It assumes both start & end=0
print(word[0:6:2]) #pto. step=2
print(word[::-1]) #nohtyp
```

```python
print(word[0:6])
print(word[0:6:1])
```

```python
print(word[3::-1])
```

```python
print("word[3:6:-1]:",word[3:6:-1])
```

```python
print(word[5::-1])
print(word[::-1])
```

```python
print(word[5:2:-1]) #it satrts at 5 and goes backward till 2
```

```python
print(word[6:2:-1]) #slicing over limit
```

```python
word = "Programming" #len = 11, last_index=10
print(word[12:3:-2])
```

```python
print(word[12:2:-2])
```

```python
word = "python"
print("word[0:6:1]:", word[0:6:1])
print("word[:6]:", word[:6]) #Both meaning same
```

```python
print(word[0:]) #prints from 0 to till end-->word[0:6:1]
print(word[3:]) #from 3 to till end --> word[3:6:1]
```

```python
print(word[:]) #word[0:6:1]
print(word[::]) #word[0:6:1]
```

```python
print(word[0:]) #0 is start_index
print(word[:5]) #5 is final_index
print(word[::1]) #1 is step
```

```python
print(word[::2])
```
