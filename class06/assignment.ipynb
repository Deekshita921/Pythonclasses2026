## String Slicing Assignement

1) Input: today is good day  
Output: yadot si doog yad

```python
sentence = "today is good day"
print(len(sentence))
```

```python
words = sentence.split()
print(words)
```

```python
for word in words:
    print(word)
```

```python
for word in words:
    print(word[::-1])
```

```python
#to make them print in same line we can use "end="
for word in words:
    print(word[::-1],end = " ")
```

2) Input: Varun Dhawan  
Output: nuraV nawahD

```python
#Generally we do
sen = "Varun Dhawan"
print(sen[::-1]) 
```

```python
wrds = sen.split() #['Varun', 'Dhawan']
for i in wrds:
    print(i[::-1], end=" ") #To give some space between words give space in between end=" "
```

## Palindrome Strings
3) Check whether the given string is palindrome or not
mom
dad

```python
Word = input("Enter the word: ").lower() #To avaoid capital letters
reverse_string = Word[::-1]
if Word == reverse_string:
    print("Is a palindrome")
else:
    print("Not a plalindrome")
```
