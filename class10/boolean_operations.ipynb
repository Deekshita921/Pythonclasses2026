```python
#True = 1
print("True     =", True)
print("True * 30 =", True*30) #1*30 = 30
print("True * 3.4 =", True * 3.4) #3.4
```

```python
#False = 0
print("False     =", False)
print("False * 30 =", False*30) #0*30 = 0
print("False * 3.4 =", False * 3.4) 
```

```python
print("True == 1 ->",True == 1) #True
print("True == 1.0 ->",True == 1.0) #True
print("True == 1.1 ->",True == 1.1) #False this is not 1
print("True == 3 ->",True == 3) #False
```

```python
print("str(1) != str(True)", str(1)!= str(True)) #True->'1' != 'True'
print("str(0) != str(False)", str(0) != str(False)) #True -> '0' != 'False'
```

```python
print("str(True * 1) == str(1)", str(True * 1) == str(1)) #'1' == '1' ->True
```

### bool() -> Built-in function

```python
#integers
# True -> non-zero
#False -> Zero
```

```python
#Strings
# True -> non-empty string
# False -> empty string
print("bool('ball'):", bool('ball'))
print("bool(' '):", bool(" ")) #White-space
print("bool(''):", bool("")) #Empty String
```

```python
print(bool(None)) #None means empty string. so False.
print(bool("None")) #not an empty string. so True.
print(bool(True))
print(bool(False))
```

```python
#braces
# True -> atleast 1 char within the braces
# False -> empty inside braces
print("(1)      :", bool((1)))
print("(1,)     :", bool((1,)))
print(bool([1]))
print(bool({1}))
print(bool({1:2}))
```

```python
print("() :", bool(()))
print("[] :", bool([]))
print("{} :", bool({}))
```

```python
#for Relational Operations
print(74 > 59)
print(bool(74 > 59))
print(74 <= 59)
print(bool(74 <= 59))
```

```python
#for logical operations
print("1 < 2 and 23 <= 23 :", 1 < 2 and 23 <= 23)
print("bool(1 < 2 and 23 <= 23):", bool(1 < 2 and 23 <= 23))
```

```python
num2 = -0.0000000000000000
print("bool(num2):", bool(num2)) #bool(0)->False
print("bool(num2 != 0):", bool(num2 != 0)) #False -> it is zero.

if num2 != 0: #bool(Flase)->False (it wont print anything as it wont go inside block)
    print("It is non-zero")
if num2: #means if (num2 != 0)->Flase
    print("It is non-zero")
if not num2 != 0: #True
    print("It is Zero")
if (
    not num2
):
    print("It is zero")
if (not num2):
    print("It is Zero")
```
