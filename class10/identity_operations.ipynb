check whether the character is vowel or consonant.
    vowel = 'aeiou' or 'AEIOU'

```python
char = input("Enter the character: ")

if(char == 'a' or char=='A' or char=='e' or char=='E' or
char=='i' or char=='I' or char=='o' or char=='O' or char=='u' or char=='U'):
    print("It is an vowel")
else:
    print("Not an vowel, It is a consonant")
```

```python
char = input("Enter the character: ").strip()

if char in ('a', 'A', 'e', 'E', 'i', 'I', 'o', 'O', 'u', 'U'):
    print(f"{char} is an vowel")
else:
    print(f"Not an vowel, {char} is a consonant")
```

```python
char = input("Enter the character: ").strip().lower()

if char in ('a','e','i','o','u'):
    print(f"{char} is an vowel")
else:
    print(f"Not an vowel, {char} is a consonant")
```
