## String Foramatting:
Types:
1.	Old Style formatting (or literal formatting)
2.	New style formatting
3.	F-strings (works from python 3.6+)


### Old Style Formatting
    %d - Integer
    %i - Integer

    %c - single char
    %s - string/char

    %f - floating point
    %F - floating point

    %o - octal
    %x - hexa decimal lowercase
    %X - hexa decimal uppercase

### %d

```python
print("My lucky num is %d")
print("My lucky num is %d" %(12))
print("My lucky num is %d" % 12)
```

```python
#if we give float value
print("My lucky num is %d" %(12.34))
print("My lucky num is %d" % 12.34)
```

```python
#if we give string
print("My lucky num is %d" %'12')
```

```python
print("My lucky num is %s" %'12') #%s for string
```

### %f

```python
print("My lucky num is %f" %12.34)
print("My lucky num is %f" %12)
print("My lucky num is %f" %'False')
```

```python
print("My lucky num is %f" % False)
print("My lucky num is %f" %True)
```

```python
print("My lucky num is %d" % False)
print("My lucky num is %d" %True)
```

```python
#float() -- 6 digits post decimal
print("%f" % 12.34)
print("%F" % 12.34)
print("%F" % 12.34223187)
print("%f" % 12.34223187)
```

```python
#float() %g -- 4-digits post decimal for large nums
#else prints same num
print("%g" % 12.34)
print("%G" % 12.34)
print("%G" % 12.342231445)
```

```python
#Exponential %e / %E
print("%e" % 12.34)
print("%E" % 12.34)
print("%E" % 12.34223123123123)
```

```python
import math

print(math.pi)          #3.141592653589793
print("%d" % math.pi)   #3
print("%f" % math.pi)   #3.141593
print("%9f" % math.pi)  # 3.141593
print("%9.3f" % math.pi) #    3.142
```

```python
print("The money %d has been credited" % 1200)
print("The money %f has been credited" % 3.1416)
print("The money %10f has been credited" % 3.1416)
print("The money %10.3f has been credited" % 3.1416)
print("The money %010.3f has been credited" % 3.1416)
#This means keep '0' instaed of spaces to fill the 10 spaces
```

### %s

```python
print("My lucky num is %s" % True) 
print("My lucky num is %s" % None)
print("My lucky num is %s" % 1)     #integer
print("My lucky num is %s" % 12.3)  #float
print("My lucky num is %s" % 133224234) #Long num
print("My lucky num is %s" % "Python") #Actual string
```

### repr() --> Raw Representation

```python
print("My lucky num is %r" % True) 
print("My lucky num is %r" % None)
print("My lucky num is %r" % 1)     #integer
print("My lucky num is %r" % 12.3)  #float
print("My lucky num is %r" % 133224234) #Long num
print("My lucky num is %r" % "Python") #Actual string
```

### %o %x %X

```python
print("%o" %34, oct(34)) #im showing the same ans using
#oct() built-in fun too for better understanding

#%x - hexadecimal lowercase
print("%x" % 34, hex(34))
#%X - hexadecimal uppercase
print("%X" % 34, hex(34))
```

```python
print("%b" %(12), bin(12))
```

### New Style String Formatting

```python
#odd braces give value as answer, even braces gives only braces no answer
print("{}".format("Uday"))
print("{{}}".format("Uday"))
print("{{{}}}".format("Uday"))
```

```python
print("{{{{}}}}".format("Uday")) #4 braces->4/2=2 braces in o/p
print("{{{{{}}}}}".format("Uday")) #5 braces->5/2=2.5->2 braces & value
```

```python
print("{{{{{{{{{{{{{{{}}}}}}}}}}}}}}}".format("Uday")) #15 braces -> 7 braces & value
```

```python
print("{} and {} are good friends".format("Uday","Kiran"))
print("I am {} years old and i pay {} tax amount".format(28,15.5))
print("I am {} years old and i pay {} tax amount".format("28","15.5"))
```

```python
print("{} and {}".format(None,True))
print("{0} and {1}".format(None, True))
print("{1} and {0}".format(None, True))
```

```python
print("{} and {}".format(None))
```

```python
print("{0} and {0}".format(None))
print("{0} and {1} are friends.{1} went to {0}'s house.".format("Uday","Ramu"))
#Here we have 4 representations & 2 values
```

```python
print("{}".format(1234567890.88))
print("{:}".format(1234567890.88))
print("{:,}".format(1234567890.88)) #comma
print("{:_}".format(1234567890.88)) #Underscore
print("{:-}".format(1234567890.88)) #minus
print("{:+}".format(1234567890.88)) #plus
print("{:+}".format(-1234567890.88)) #Giving plus+ for negative num.It wont work.
```

### Filling in new style

```python
print("{:=10}".format(12348)) #     12345-->5spaces followed by 5 nums = 10 digit space
print("{:0=10}".format(12348)) #0000012348 -->5 0's followed by 5 nums = 10 digit space
print("{:a=10}".format(12348)) #aaaaa12348
```

### Alignment 
    -Like we done in strings we used "center", "ljust", "rjust".
    similarly we can do it here as well. 
    Not only for strings but for everything we can do now.

```python
print("{:^20}".format("Centered")) #similar to center(20)
print("{:<20}".format("Left aligned")) #ljust(20)
print("{:>20}".format("Right aligned")) #rjust(20)
```

```python
print("{:-^20}".format("Centered")) #center(20,"-")
print("{:=^20}".format("Centered")) #center(20,"=")
```

```python
print(f"{'quoted string'}")
print(f"{"Quoted string"}")

#to get braces in string
print(f"{{"quoted string"}}") #{quoted string}
```

```python
print(f"{'quoted string'}")
print(f"{"Quoted string"}")

#to get braces in string
print(f'{"quoted string"}') #quoted string
print(f'{{"quoted string"}}') #{"quoted string"}
print(f'{{{"quoted string"}}}') #{quoted string}
#Here it didn't gave quotes in answer which is correct because
#if we give odd no.of quotes it gives answer with braces
#even no.of quotes gives just what is written as it is

print(f"{4 * 10}") #40
print(f"{{4 * 10}}") #{4 * 10} ->even no.of braces
print(f"{{{4 * 10}}}") # {40}
print(f"{{{{4 * 10}}}}") #{{4 * 10}}
print(f"{{{{{4 * 10}}}}}") #{{40}}
```
