### 2. Caesar Cipher
    implementing a Caesar Cipher with a shift of +3.

    In a Caesar cipher, every letter is shifted forward by 3 positions in the alphabet.
    Ex:-
    egg -> hjj
    bindu -> elgqx
    yash -> bdvk

    #Assumption: Ignore case-sensitivity
    HINTS: ord(), chr(), %

```python
word = input("Enter the word: ")
shift = 3
result = ""
for ch in word:
    if ch.isalpha():
        new_ch = chr((ord(ch) - ord('a') + shift) % 26 + ord('a'))
        result += new_ch
    else:
        result += ch
print(f"encrypted word for {word}= ", result)
```
