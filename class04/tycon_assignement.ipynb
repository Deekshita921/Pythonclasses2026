# Type Conversion Assignment

## Binary -> octal

```python
bin(23)
```

```python
oct(23)
```

```python
#first let us convert bin->int and then convert int->oct known process
oct(int(bin(23),base=2))
```

```python
print(oct(int(bin(23), base=2)))
print(oct(int("0b10111", base=2)))
print(oct(int("10111",base=2)))
print(oct(int("0b10111", base=0)))
```

```python
#or we can do
a = "0b10111" #Binary format of 23
#now convert it to oct
oct(int(a, base=2))
```

## binary -> hexadecimal

```python
hex(23)
```

```python
hex(int(bin(23), base=2))
```

```python
a = "0b10111" #23
hex(int(a, base=2))
```

## Octal -> Binary

```python
oct(9)
```

```python
a = "0o11"
bin(int(a, base=8)) #since inside it is octal give base=8
```

```python
#check whether the got answer "0b1001" is 9 or not
int("0b1001", base=2)
```

## Octal -> Hexadecimal

```python
a = "0o11"
hex(int(a, base=8))
```

```python
#check
int("0x9", base=16)
```

## Hexadecimal -> binary

```python
hex(23)
```

```python
a = "0x17" #23
bin(int(a, base=16))
```

```python
#check
int("0b10111", base=2)
```

## Hexadecimal -> Octal

```python
a = "0x17" #23
oct(int(a, base=16))
```
