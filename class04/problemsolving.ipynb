## Practical Problem Solving

1.Ration Store Problem
----------------------------------------------------------------------
item            cost            quantity            Amount
----------------------------------------------------------------------
wheat           25/kg           30 kgs          25 * 30 = 750/-
rice            12/kg           50 kgs          12 * 50 = 600/-
                                            --------------------------
                                                          1350/-
                                            subsidy 20%   -270/-
                                            --------------------------
                                            Billable Amount : 1080/-

```python
#Ration Store
cost_of_wheat = 25
cost_of_rice = 12

quan_of_wheat = 30
quan_of_rice = 50

sp_of_wheat = cost_of_wheat * quan_of_wheat
sp_of_rice = cost_of_rice * quan_of_rice 

total_sp = sp_of_wheat + sp_of_rice
print("Total Selling Price: ", total_sp)

subsidy_amount = (total_sp * 20)/100
print("Subsidy amount =", subsidy_amount)

billable_amount = total_sp - subsidy_amount
print("Total Billable Amount =", billable_amount)
```

2.Fruit Store

```python
price_of_apple = 12
price_of_mangos = 34

quan_of_apple = 5 * 12
quan_of_mangos = 3 * 12

sp_of_apple = price_of_apple * quan_of_apple
sp_of_mangos = price_of_mangos * quan_of_mangos

total_sp_fruitstore = sp_of_apple + sp_of_mangos
print("Total Selling Price =", total_sp_fruitstore)

discount = (total_sp_fruitstore * 2)/100
print("Discount =", discount)

amount_after_discount = total_sp_fruitstore - discount
print("Total Amount after discount =", amount_after_discount)

GST = (amount_after_discount * 12.5)/100
print("GST =", GST)

Billable_amount = amount_after_discount + GST
print("Billable amount of fruit store =", Billable_amount)
```
