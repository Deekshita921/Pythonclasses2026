```python
# int -> Float
print(12, float(12))
```

```python
#int -> string
12, str(12)
```

```python
print(12, str(12))
```

```python
# Float -> int
print(3.14, int(3.14))

#Float -> String
print(3.14, str(3.14))
```

```python
#String to Int
print("23", int("23"))
print("23", int("23 "))
print("23", int("   23"))
```

```python
print("23", int("2 3"))
```

```python
#string to int
int("23.2")
```

```python
#String to Float
print("23", float("23"))
print("23 ", float("23"))
print("23.24", float("23.24"))
print("23. ", float("23. "))
#print("2 3. ", float("2 3. ")) #ValueError: Could not convert str to float: "2 3. "
#space in between so cant convert
```

```python
print(float()) #0.0
print(float("+1.23")) #1.23
print(float(+1.23)) #1.23
print(float("-1.23")) #-1.23
print(float(" -1.23\n")) #-1.23 #removes spaces and any escape chars
print(float("1e-003")) #0.001
print(float("+1E6")) #1000000.0
print(float("-Infinity")) #-inf str->float
#print(float(-Infinity)) #NameError "name Infinity is not defined"
print(float("-inf")) #-inf str->float
print(float("+inf")) #inf str->float
```

```python
#Converting to string
#Anything can be converted to string but not vice verse

print("str(12): ", str(12))
print("str(12.3333): ", str(12.3333))
print("str(-0.000012): ", str(-0.000012))
print("str(True): ", str(True))
print("str(None): ", str(None))
```

## Decimal -> Octal

```python
print(oct(9))
print(oct(-23))
```

## Octal -> Decimal

```python
print(int(oct(9), base=8))
print(int("0o11", base=8)) #has prefix"0o" and base
print(int("0o11",base=0)) #has prefix & base=0
print(int("11", base=8)) #no prefix, has base
print(int("11",8)) #we can write base=8 or just 8
print(int("11")) #no prefix & no base
print(int("11",0)) #won't work
```

## Decimal -> Hexadecimal

```python
hex(23)
```

## Hexadecimal -> Decimal

```python
print(int(hex(23), base=16))
print(int("0x17", base=16)) #has prefix & base
print(int("0x17", base=0)) #has prefix & base=0(means no base)
print(int("17", base=16)) #no prefix & has base
print(int("17",0)) #no prefix & base similarlly int("17")
```

Finally

```python
int(bin(9), base = 2) #binary -> int (or) decimal
```

```python
int(oct(9), base = 8) #octal -> int
```

```python
int(hex(9), base = 16) #hexadecimal -> int
```

## is_integer()
    checks whether the num is int or not. If int returns "True" else "False"

```python
print((2.0).is_integer())
print((2.1).is_integer())
print((-2.1).is_integer())
```

## isinstance()

```python
isinstance(2.0,int)
```

```python
isinstance(2.0,float)
```

```python
print(isinstance(2,int)) #True
print(isinstance(2.0,float)) #True
print(isinstance("Hi", str)) #True
print(isinstance([1,2,3], list)) #True
print(isinstance({"name":"Deeksha"}, dict)) #True

# we can even check for multiple data types
print(isinstance(2, (int,float))) #True
```
