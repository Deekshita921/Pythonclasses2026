# Absolute Number 
# abs()

```python
#positive number
abs(3)
```

```python
#Negative number
abs(-3)
```

```python
print(abs(3)) #positive number
print(abs(-3)) #Negative number
print(abs(0)) #Zero
print(abs(3.1)) #Float Value
print(abs(-3.1))
```

```python
#Complex Numbers
print(abs(-3+4j))
print(abs(3+4j))
```

```python
import math

print(math.sqrt(pow(-3,2) + pow(4,2))) #5.0
print(math.sqrt((-3) ** 2 + 4**2))     #5.0
```

# Complex Numbers

```python
num1 = 2 + 3j
print(num1, type(num1))
```

```python
num1 = -2 + 3j
print(num1, type(num1))
```

```python
num1 = -0.02 + 3.234j
print(num1, type(num1))
```

```python
num1 = -0.2 + 0j #Even if we keep 0 for j. Still it is complex because j is there
print(num1, type(num1))
```

```python
num1 = -0.0j #if there is a j then it is complex number
print(num1, type(num1))
```

```python
num1 = 4*j 
print(num1, type(num1))
```

```python
num1 = 4 * 1j #4 * 1j only becomes 4j not 4 * j
print(num1, type(num1))
```

```python
num1 = j4
print(num1, type(num1))
```

Note: 4*j,j*4,j4 are not possible
