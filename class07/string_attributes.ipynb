### Assignment: Modify the partition attribute to specific index positions

partition() does not modify the string, and it does not work with indexes. It only splits a string based on a separator and returns a tuple of 3 elements.
    String is immutable in python and partition gives result in tuple which is also immutable.
    -we can modify it by converting the tuple to list. Since list is mutable.

```python
text = "mango-apple-banana"
result = list(text.partition("-")) #('mango','-','apple-banana') -> ['mango','-','apple-banana]
print("Result before modification: ",result)
result[0] = 'orange'
print("Result after modification: ",result)
```

### Assignment: Replace from right to left. In reverse direction.

```python
#First from left to right
#Replace 2 with 3
p = "2221222".split("2",2)
print("Split parts:", p)
result = "3".join(p) #"3".join(['','','','21222'])
print("Result:", result) #"3321222"
```

```python
#From right to left
parts = "2221222".rsplit("2",2)
print("Split parts:", parts)
resul = "3".join(parts) #"3".join(['22212','',''])
print("Result:", resul)
```
